# 時間測量

使用`timeit`模組當中的`timeit`函式，格式為 (欲測量的函式, 環境建置, number=重複執行次數)，且前兩參數為字串格式<br>
重複執行次數預設為一百萬次`number=1000000`，回傳數值單位為**秒**；也可以使用`timeit.Timer`來分段程式碼

```python
import timeit

def A(alist, item):
    return item in alist
def B(adict, item):
    return item in adict

alist = [x for x in range(1000)]
adict = {y:0 for y in range(1000)}
item = 999

print('List Search:', timeit.timeit('A(alist,item)', 'from __main__ import A, alist, item', number=1000))
print('List Search:', timeit.timeit('A(alist,item)', 'from __main__ import A, alist, item'))
print('Dict Search:', timeit.timeit('B(adict,item)', 'from __main__ import B, adict, item'))

t = timeit.Timer('A(alist,item)', 'from __main__ import A, alist, item')
print('List Search:', t.timeit(number=1000))
```
